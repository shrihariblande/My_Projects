### 1.PROBLEM STATEMENT:

As the Rapidly growing population, the demand of drinking water is increased and due to contamination or unsafe access of drinking water it leads to Waterborne diseases, then it will increase burden on the Health Care as well as finance domain, so it needs to predict the drinking of water is potable or not before the supply of water to community.

### 2.Data Gathering

```python
# Importing Libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler , StandardScaler
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing Data
df = pd.read_csv("N_Drinking_Water_Quality.csv")
df.head(10)
```

### Exploratory Data Analysis

```python
df.drop(["Index","Unnamed: 0"],inplace = True, axis = 1)
```

```python
df.describe()
```

```python
# Detecting null values 
df.isna().sum()
```

```python
# percent of null values
percentage_missing = df.isna().sum()*100/len(df)
percentage_missing
```

```python
df.info()
```

```python
# Preplacing object dtype into int
df["Color"].unique()
```

```python
df["Source"].unique()
```

```python
df["Month"].unique()
```

```python
df["Color"].replace({'Colorless': 0,'Near Colorless': 1,"Faint Yellow":2,"Yellow":3,"Light Yellow":4},inplace=True)
```

```python
df["Source"].replace({'Well': 0,'Lake': 1,"Aquifer":2,"Stream":3,"Ground":4,"Spring":5,"Reservoir":6,"River":7},inplace=True)
```

```python
df["Month"].replace({'March': 0,'October': 1,"May":2,"April":3,"June":4,"July":5,"September":6,"November":7,'August':8,'December':9,'January':10,'February':11},inplace=True)
```

```python
df.info()
```

```python
# outlier Detecting
q1 = df['pH'].quantile(0.25)
q2 = df['pH'].quantile(0.5)
q3 = df['pH'].quantile(0.75)

print('Q1 :' , q1)
print('Q2 :' , q2)
print('Q3 :' , q3)
print('min :',df['pH'].min() )
print('max :',df['pH'].max() )
print('mean :',df['pH'].mean() )


iqr = q3 - q1
print('IQR :',iqr)

# upper_tail = q3 + 1.5*iqr
upper_tail = df['pH'].quantile(0.97)

lower_tail = q1 - 1.5*iqr
print('Upper_tail :',upper_tail)

print('count_of_outliers :',len(df['pH'].loc[df['pH']>upper_tail]))
```

### Feature Engineering

```python
# Handling Missing Values
df["pH"].fillna(df["pH"].median(),inplace = True)
df["Iron"].fillna(df["Iron"].median(),inplace = True)
df["Nitrate"].fillna(df["Nitrate"].median(),inplace = True)
df["Chloride"].fillna(df["Chloride"].median(),inplace = True)
df["Lead"].fillna(df["Lead"].median(),inplace = True)
df["Zinc"].fillna(df["Zinc"].median(),inplace = True)
df["Color"].fillna(df["Color"].mode()[0],inplace = True)
df["Turbidity"].fillna(df["Turbidity"].median(),inplace = True)
df["Fluoride"].fillna(df["Fluoride"].median(),inplace = True)
df["Copper"].fillna(df["Copper"].median(),inplace = True)
df["Odor"].fillna(df["Odor"].median(),inplace = True)
df["Sulfate"].fillna(df["Sulfate"].median(),inplace = True)
df["Conductivity"].fillna(df["Conductivity"].median(),inplace = True)
df["Chlorine"].fillna(df["Chlorine"].median(),inplace = True)
df["Manganese"].fillna(df["Manganese"].median(),inplace = True)
df["Total Dissolved Solids"].fillna(df["Total Dissolved Solids"].median(),inplace = True)
df["Source"].fillna(df["Source"].mode()[0],inplace = True)
df["Water Temperature"].fillna(df["Water Temperature"].median(),inplace = True)
df["Air Temperature"].fillna(df["Air Temperature"].median(),inplace = True)
df["Month"].fillna(df["Month"].mode()[0],inplace = True)
df["Day"].fillna(df["Day"].median(),inplace = True)
df["Time of Day"].fillna(df["Time of Day"].median(),inplace = True)
```

```python
# Scaling data
x = df.drop("Target",axis = 1)
y = df["Target"]
```

```python
# minmax Scaler (Normalization)
normal_scaler = MinMaxScaler(feature_range = (0,1))
x_scaled = normal_scaler.fit_transform(x)
```

```python
x_normal_df = pd.DataFrame(x_scaled,columns = x.columns)
x_normal_df
```

```python
# Standardization
std_scaler = StandardScaler()
x_scaled = std_scaler.fit_transform(x)
```

```python
x_std_df = pd.DataFrame(x_scaled,columns = x.columns)
x_std_df
```

```python
# Handling Outliers
```

```python
# Splitting the data in to train test split
x_train,x_test,y_train,y_test = train_test_split(x_std_df,y,test_size = 0.2,random_state = 1,stratify=y)
```

### Logistic Regression Model

```python
# Model Instantiating
log_model = LogisticRegression()
```

```python
# Predicting model
y_pred = log_model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

### Decision tree Model

```python
dt_model = DecisionTreeClassifier(criterion = 'gini')
dt_model.fit(x_train,y_train)
```

```python
y_pred_test = dt_model.predict(x_test)


testing_accuracy = accuracy_score(y_test,y_pred_test)
print('Testing data accuracy is :', testing_accuracy)
print()
clf_report = classification_report(y_test,y_pred_test)
print('Classification report is\n :',clf_report)
print()
confusion_matrix(y_test,y_pred_test)
```

```python
# trainning Accuracy
```

```python
y_pred_train = dt_model.predict(x_train)


training_accuracy = accuracy_score(y_train,y_pred_train)
print('Train data accuracy is :', training_accuracy)
print()
clf_report = classification_report(y_train,y_pred_train)
print('Classification report is\n :',clf_report)
print()
confusion_matrix(y_train,y_pred_train)
```

```python
# Hyperparameter tunning
```

```python
# randomizedSearchCV
hyperparameters = {'criterion' : ['gini','entropy'],
                   'max_depth' : np.arange(2,10),
                   'min_samples_split' : np.arange(2,15),
                   'min_samples_leaf' : np.arange(1,15)}

dt_model = DecisionTreeClassifier()
rscv_dt_model = RandomizedSearchCV(dt_model,hyperparameters,cv = 5)
rscv_dt_model.fit(x_train,y_train)
```
